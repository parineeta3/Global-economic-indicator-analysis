```python
country="Zimbabwe"
gdp_growth_data = data[data['country_name']==country][['year','GDP Growth (% Annual)']]
print(gdp_growth_data)
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='year', y='GDP Growth (% Annual)', data=gdp_growth_data)
plt.title(f"GDP Growth Trend in {country} (2010-2025)")
plt.xlabel("Year")
plt.ylabel("GDP Growth (% Annual)")
plt.grid(True)
plt.show()
```

```python
country="India"
gdp_growth_data = data[data['country_name']==country][['year','GDP Growth (% Annual)']]

plt.figure(figsize=(8, 6))
sns.barplot(x='year', y='GDP Growth (% Annual)', data=gdp_growth_data)
plt.title(f"GDP Growth Trend in {country} (2010-2025)")
plt.xlabel("Year")
plt.ylabel("GDP Growth (% Annual)")
plt.show()
```

```python
country="Germany"
gdp_growth_data = data[data['country_name']==country][['year','GDP Growth (% Annual)']]


plt.figure(figsize=(8, 6))
sns.scatterplot(x='year', y='GDP Growth (% Annual)', data=gdp_growth_data)
plt.title(f"GDP Growth Trend in {country} (2010-2025)")
plt.xlabel("Year")
plt.ylabel("GDP Growth (% Annual)")
plt.grid(True)
plt.show()
```

```python
##UNEMPLOYEMENT COMPARISON ACROSS COUNTRIES##

avg_unemployment = data.groupby('country_name')['Unemployment Rate (%)'].mean().nlargest(5)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_unemployment.index,y=avg_unemployment.values)
plt.title("AVERAGE UNEMPLOYMENT  RATE (TOP 5 COUNTRIES)")
plt.xlabel("country")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
avg_unemployment = data.groupby('country_name')['Unemployment Rate (%)'].mean().nsmallest(5)
print(avg_unemployment)
plt.figure(figsize=(8,5))
sns.barplot(x=avg_unemployment.index,y=avg_unemployment.values)
plt.title("COUNTRIES WITH LOWEST AVERAGE UNEMPLOYMENT  RATE ")
plt.xlabel("country")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
# Correlation matrix
indicators = ['GDP (Current USD)', 'Inflation (CPI %)', 'Unemployment Rate (%)', 
              'Interest Rate (Real, %)', 'GDP Growth (% Annual)']
correlation_data = data[indicators].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title("Correlation Between Key Economic Indicators")
plt.show()
```

```python
# Summarize indicators by year
summary = data.groupby('year')[['Tax Revenue (% of GDP)', 'Government Expense (% of GDP)']].mean()

summary.plot(kind='bar', figsize=(12, 6))
plt.title("Tax Revenue and Government Expense (% of GDP) by Year")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.grid(axis='y')
plt.show()
```

```python
### TOP 5 COUNTRIES WITH HIGHEST AVERAGE GDP GROWTH ###
top_gdp_countries= data.groupby('country_name')['GDP Growth (% Annual)'].mean().nlargest(5)

plt.figure(figsize=(10,6))
sns.barplot(x=top_gdp_countries.index, y= top_gdp_countries.values, palette="viridis")
plt.title("AVERAGE GDP GROWTH %")
plt.xlabel("country")
plt.ylabel("gdp growth")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()

```

```python
### BOTTOM 5 COUNTRIES WITH LOWEST AVERAGE GDP GROWTH ###
bottom_gdp_countries= data.groupby('country_name')['GDP Growth (% Annual)'].mean().nsmallest(5)

plt.figure(figsize=(10,6))
sns.barplot(x=bottom_gdp_countries.index, y= bottom_gdp_countries.values, palette="viridis")
plt.title("AVERAGE GDP GROWTH %")
plt.xlabel("country")
plt.ylabel("gdp growth")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```
